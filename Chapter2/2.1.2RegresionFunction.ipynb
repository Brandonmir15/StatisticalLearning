# Part 2 of our Regression Function


In the previous notebook, we were given a set of input output pairs (洧논,洧녽).Our goal was to build a simple model that can predict a value of 洧녧 for any given 洧논.

To do this, we used a simple form of regression
we predicted the output 洧녧 at a certain value of 洧논 by taking the average of all the observed 洧녧 values that occur at that same 洧논.

Which is Represented by this formula:

$$
\hat{f}(x_0)
= \frac{1}{n}
\sum_{\substack{i \\ x_i = x_0}} y_i
$$


There is one problem, which is in many data sets if we are trying to predict an output at a given 洧논 there might not be enough or any observed Y values for that 洧논 that we can find the average of. So what do we do here?

So In this case we will find the average of all the observed Y values that fall in the neighborhood of x not just only on x.

This is shown by the formula below where N(X) represents the neighborhood around x

$$
\hat{f}(x) = \mathrm{Ave}\bigl( Y \mid X \in \mathcal{N}(x) \bigr)
$$
