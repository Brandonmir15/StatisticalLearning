# K Nearest Neighbors

K Neaserst Neighbors is a model for predicting the outcomes for classification models

The way that K nearest neighbors works is actually kinda intuitive. if we want to predict the output of a certain x what we can do is get x's k nearest neighbors. Here you can set k to would you would like so for example if we set k to 5, what we would do is take our x's 5 nearst neighbors and based off of them we can predict what our output should be


Say we were given a plot like the one above. and we wanted to find the predict the value that will be at $C(x_1, x_2)$ at $x_1 = 1, x_2 = 1$. we will make $k=5$. What we do here is we will go to the 5 closest neighbors and see weather they are yellow or blue. From the looks of it there seem to be 4 Blue neighbors and 1 yellow. So there for we will predict $\hat{C}(x_1, x_2) = Blue$ 

```python
import math

# here we are going to be given a list of points, where each is an observed data point represented as a tuple
# each of points[i] corresponds to the value values[i] 
# for right now we will assume the prediction point is one point
# for this i know that sorting is not the optimal approach but we will do it for simplicity sake

points = [ (1.0, 2.0), (2.0, 1.0), (3.0, 3.0), (5.0, 4.0), (6.0, 5.0), (7.0, 3.0), (8.0, 4.0), (2.0, 4.0), (1.5, 3.5), (3.5, 2.0)]

values = [ "yellow", "yellow", "yellow", "yellow", "blue", "blue", "blue", "yellow", "yellow","blue"]

prediction_point = [(1.0, 4.2), (9.0, 4.0), (7.8, 4.7), (3.8, 2.6), (1.2, 2.8)]


def KNN(points:list, values:list, predictionPoints:list, k=5) -> any:

    results = []

    for eachPredictionPoint in predictionPoints:

        frequencyDict = {}
        distances = []

        for i in range(len(points)):

            distance = abs(math.dist(points[i], eachPredictionPoint))
            distances.append((distance,i))  # make sure to keep the index in order to get the value later
            # get the distances between every point and the prediction point
        

        distances.sort() # sort our values so we can grab the closet k neighbors
        nearestNeighbors = distances[:k+1] # these are our k neaerst neighbors
        nearestNeighborsValues = []

        

        for eachPoint in nearestNeighbors:
            nearestNeighborsValues.append(values[eachPoint[1]])

    
        for eachValue in nearestNeighborsValues:
            if eachValue not in frequencyDict:
                frequencyDict[eachValue] = 1
            else:
                frequencyDict[eachValue] += 1
        
        
        maxFreq = 0 # TODO: fix logic here
        result = None

        for eachItem in frequencyDict:
            if frequencyDict[eachItem] > maxFreq:
                maxFreq = frequencyDict[eachItem]
                result = eachItem


        results.append(result)
    
    return results

if __name__ == '__main__':
    print("Point Predictions: " + str(KNN(points, values, prediction_point)))

    





```
