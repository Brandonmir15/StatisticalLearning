### Below is the mean square error function

$$
\text{MSE}
= \frac{1}{n}
\sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$


```python
# for right now I will just impliment the Mean sqared error function

y_true = [3, 5, 2, 7, 4]
y_predicted = [2, 5, 4, 6, 5]

# i - will be interpred as x0 and 

def meanSquaredError(trueValues:list, predictedValues:list) -> float:
    
    vals = []
    for i in range(len(predictedValues)):
        
       error = trueValues[i] - predictedValues[i]

       squaredError = error ** 2

       vals.append(squaredError)
    

    mse = round(sum(vals) / len(predictedValues), 2)

    return mse

       


if __name__ == '__main__':
    print(meanSquaredError(y_true,y_predicted))


```
