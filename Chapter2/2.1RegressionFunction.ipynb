So our model will be written as 
$$

## Creating Predictions

Given this function or a function f for example what would be a good prediction for a value such as f(4) and how would we go about predicting it

One Good method to go about predicting this would be to take the avrage of all the values at f(4) and ploting it.

This is what is known as a regression function. now lets write some code for this

```python
"""
Given a function f why countains Y(response) values. we want to predict what a good value would be for f(x) 
Here is a very simple implimentation of a regression function where we will take the given Y values at that point x and find their average to get a response

For this Example we are going to store points using a ditinary to keep things simple

The dict will be layed at as below with the key of the dict being the x value
and the value being an array of the Y values that that x value contains
points {1: [3,5],
        2: [5,6,7]}

"""


def simpleRegression(points:dict) -> list:

    # Remember to find a good value for f(x) we would want to find the average of the Y's at that x value (for this example)
    # We will start off by itterating over everypoint in our Dict geting its Y values and averaging them

    result = []
    for eachPoint in points:
        
        localSum = sum(points[eachPoint]) 
        average = (localSum / len(points[eachPoint]))
        result.append(round(average,2))

    return result


if __name__ == '__main__':

    points = {1:[3,5],
          2:[5,6,7],
          3:[7,9,10],
          4:[10,11,13],
          5:[11,13,14]}

    print(simpleRegression(points))
```
