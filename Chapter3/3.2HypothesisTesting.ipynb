# Hypothesis Testing

The most common Hypothesis test involves testing the null hypothesis

$H_0$: There is no relationship between $X$ and $Y$

$H_A$: There is some relationship beween X and Y

Thinking about it in terms of math

$H_0$: $\beta_1 = 0$ 

$H_A$: $\beta_1 \neq 0$ 

in order to test the null hypothesis we will compute the t-statistic, which is $$t = \frac{(\hat{\beta_0} - 0)}{SE(\hat{\beta_1})} $$

Where SE is $$\frac{\sigma^2}{\sum_{i=1}^n(x_i - \overline{x})^2}$$

and $\sigma^2$ is the $var(\epsilon)$

To Assess the overall accuracy of our model we an find the Residual Standard Error which is found via the equation $$\sqrt{\frac{1}{n-2}RSS}$$

Lastly, to find out $R^2$ value we will do $\frac{TSS - RSS}{TSS}$

I know that this is a lot of equations, but lets impliment them now and create a kind of mini librarys so that we can easily use them in the future
