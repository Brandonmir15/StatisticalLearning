# Simple Linear Regression

Linear regression is a simple way to conduct supervise learning and it assumes that the dependence of $Y$ on $X_1,X_2...X_n$ is linear

Simple Linear Regression uses a single predictor $X_1$

Simple Linear Regression Model: $$ Y = \beta_0 + \beta_1X + \epsilon $$

Where $\beta_0$ and $\beta_1$ are unknown constants that represent the intercept and the slope

Given that we are going to be predicting these values our formula will look like: $$ \hat{y} = \hat{\beta_0} + \hat{\beta_1}X + \epsilon  $$

## How are we going to predict our values for this constants?

What we can do here first is make a guess and then find the error of our guess $$e_i = y_i - \hat{y}_i$$

here $e$ is our error $y_i$ is the true value and $\hat{y}_i$ is our guess

if we square all of our errors we will get what is called the Residual Sum of Square (RSS) $$ RSS = e_1^2 + e_2^2 +... e_n^2$$

Once we have this we can use an approach called the least squares approach which will find us a $\hat{\beta_0}$ and $\hat{\beta_1}$ that minimizes our RSS

The approach is calculated as below: $$\hat{\beta_1} = \frac{\sum_{i=1}^n{(x_i-\overline{x})(y_i-\overline{y})}}{\sum_{i=1}^n{(x_i-\overline{x})}}  $$ $$ \beta_0 = \overline{y} - \beta_1X$$

These Equations are derrived through partial diffrentiation: link to a good explination here: https://www.youtube.com/watch?v=ewnc1cXJmGA&list=LL&index=2

## Lets Impliment Simple Linear Regression !!!!!!!!!!!!!!!!!!

```python

#For this simple Linear regression we are going to  return a tuple of our slope and intercept (slope,intercept) 
# we will impliment the mean calculation to

# we also assume these are parallel arrays that we are intaking

x = [1, 3, 5, 7, 9, 11]
y = [10, 15, 22, 26, 30, 36]


def simpleLinearReggresion(inputs, outputs) -> tuple:
    
    x_mean = avg(inputs)
    y_mean = avg(outputs)
    numerator = []
    denom = []

    for i in range(len(inputs)):

        diffX = inputs[i] - x_mean
        diffY = outputs[i] - y_mean
        prod = diffX * diffY
        numerator.append(prod)
        denom.append(diffX ** 2)


    
    numeratorSum = sum(numerator)
    denomSum = sum(denom) 

    slope = numeratorSum / denomSum

    intercept = y_mean - (slope * x_mean)

    return (slope,intercept)




def avg(nums:list) -> float:
    return sum(nums)/len(nums)


if __name__ == '__main__':
    print(simpleLinearReggresion(x,y))



```
